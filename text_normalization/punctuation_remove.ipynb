```python
with open("data.txt", "r") as f:
    data = f.read()

punctuation = "!#$%&'()*+,-./:;<=>?@[\]^_`{|}~"

result = ""

for ch in data:
    if ch not in punctuation:
        result += ch

print(result)
```
