```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
with open(r"D:\Raj\Learning\genai\text_normalization\data.txt", "r") as file:
    data = file.read()
```

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(data)
```

```python
updated_token = []
for word in tokens:
    if word not in stop_words:
        updated_token.append(word)
print(updated_token,end=" ")
```
